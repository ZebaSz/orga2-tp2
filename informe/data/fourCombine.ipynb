```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import random
import glob
import os
```

```python
dfO0 = pd.read_csv('fourCombine-control.csv')

dfO3 = pd.read_csv('fourCombine-O3.csv')
dfO3.set_index(['Algoritmo'], inplace=True)
dfO3.drop(1, inplace=True)
dfO3.reset_index(inplace=True)
dfO3['Algoritmo'] = dfO3['Algoritmo'].map({0: 2})

df = pd.concat([dfO0, dfO3], ignore_index=True)
df['Algoritmo'] = df['Algoritmo'].map({1: 'ASM - SSE', 0: 'C - O0', 2: 'C - O3'})

#df = df.groupby(['Algoritmo', 'Tamao']).min()
df.reset_index(inplace=True)

sns.lmplot(x='Tamao', y='Ticks', data=df, hue='Algoritmo', truncate=True, size=7, aspect=1.4)
plt.ylabel('Tiempo en ticks')
plt.gca().get_yaxis().get_major_formatter().set_scientific(False)
plt.show()
```

```python
dfO0 = pd.read_csv('fourCombine-control.csv')

dfO3 = pd.read_csv('fourCombine-unroll.csv')
dfO3.set_index(['Algoritmo'], inplace=True)
dfO3.drop(0, inplace=True)
dfO3.reset_index(inplace=True)
dfO3['Algoritmo'] = dfO3['Algoritmo'].map({1: 2})

df = pd.concat([dfO0, dfO3], ignore_index=True)
df['Algoritmo'] = df['Algoritmo'].map({1: 'ASM - SSE', 0: 'C - O0', 2: 'ASM - SSE - UNROLL'})

df = df.groupby(['Algoritmo', 'Tamao']).min()

df.reset_index(inplace=True)

sns.lmplot(x='Tamao', y='Ticks', data=df, hue='Algoritmo', truncate=True, size=7, aspect=1.4)
plt.ylabel('Tiempo en ticks')
plt.gca().get_yaxis().get_major_formatter().set_scientific(False)
plt.show()
```

```python
df1CU = pd.read_csv('fourCombine-1ciclo.csv')

df1C = pd.read_csv('fourCombine-1ciclo-unroll.csv')
df1CU.set_index(['Algoritmo'], inplace=True)
df1CU.reset_index(inplace=True)
df1CU['Algoritmo'] = df1CU['Algoritmo'].map({1: 2})
df = pd.concat([df1C, df1CU], ignore_index=True)
df['Algoritmo'] = df['Algoritmo'].map({0:'C O1', 1: 'ASM - SSE - UNROLL', 2: 'ASM - SSE'})
df = df.groupby(['Algoritmo', 'Ciclo']).mean()
df.reset_index(inplace=True)
sns.barplot(x='Ciclo', y='Ticks', data=df, hue='Algoritmo')
plt.show()
```
