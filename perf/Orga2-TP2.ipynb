```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import random
import glob
import os
```

```python
dfrgb2yuv = pd.read_csv('build/rgb2yuv.csv')

plt.clf()
plt.figure(figsize=(12, 6))
sns.factorplot(x='Tamaño', y='Ticks', data=dfrgb2yuv, hue='Algoritmo')

plt.ylabel('Tiempo en ticks')
plt.gca().get_yaxis().get_major_formatter().set_scientific(False)

plt.show()
```

```python
plt.clf()
plt.figure(figsize=(12, 6))
sns.barplot(x='Tamaño', y='Ticks', data=dfrgb2yuv, hue='Algoritmo')

plt.ylabel('Tiempo en ticks')
plt.gca().get_yaxis().get_major_formatter().set_scientific(False)

plt.show()
```
